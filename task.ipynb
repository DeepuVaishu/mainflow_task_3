```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Bar Chart: Top 10 channels by total views

```python
top_channels = data1.groupby('channel_title')['views'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
plt.bar(top_channels.index, top_channels.values, color='skyblue')

plt.title('Top 10 Channels by Total Views',fontsize=16,color='maroon')
plt.xlabel('Channel Title',fontsize=12,color='red')
plt.ylabel('Total Views',fontsize=12,color='red')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### 2. Line Chart: Trend of likes, dislikes, and comment counts over time for a selected video

```python
# Selecting a popular video by the highest number of views
selected_video = data1.loc[data1['views'].idxmax(), 'title']
selected_video_data = data1[data1['title'] == selected_video].sort_values(by='trending_date')

plt.figure(figsize=(12, 6))
plt.plot(selected_video_data['trending_date'], selected_video_data['likes'], label='Likes', marker='o')
plt.plot(selected_video_data['trending_date'], selected_video_data['dislikes'], label='Dislikes', marker='o')
plt.plot(selected_video_data['trending_date'], selected_video_data['comment_count'], label='Comment Count', marker='o')
plt.title(f'Engagement Trend for Video: "{selected_video}"',fontsize=16,color='navy')
plt.xlabel('Trending Date',fontsize=12,color='blue')
plt.ylabel('Count',fontsize=12,color='blue')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### 3. Scatter Plot: Relationship between views and likes

```python
plt.figure(figsize=(10, 6))
plt.scatter(data1['views'], data1['likes'], alpha=0.5, color='purple')
plt.title('Scatter Plot of Views vs. Likes',fontsize=16,color='darkblue')
plt.xlabel('Views',fontsize=12,color='darkcyan')
plt.ylabel('Likes',fontsize=12,color='darkcyan')
plt.xscale('log')  # Applying logarithmic scale for better visualization
plt.yscale('log')
plt.grid(True)
plt.show()
```

### 4. Histogram: Distribution of views, likes, and dislikes

```python

plt.figure(figsize=(12, 8))

# Views Histogram
plt.subplot(3, 1, 1)
plt.hist(data1['views'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.xscale('log')

# Likes Histogram
plt.subplot(3, 1, 2)
plt.hist(data1['likes'], bins=50, color='lightgreen', edgecolor='black')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.xscale('log')

# Dislikes Histogram
plt.subplot(3, 1, 3)
plt.hist(data1['dislikes'], bins=50, color='salmon', edgecolor='black')
plt.title('Distribution of Dislikes')
plt.xlabel('Dislikes')
plt.ylabel('Frequency')
plt.xscale('log')

plt.tight_layout()
plt.show()
```

### 5. Plotting a Box Plot for views, likes, and dislikes using the quantiles

```python
# Selecting the relevant columns
metrics = data2[['views', 'likes', 'dislikes']]

plt.figure(figsize=(10, 6))
sns.boxplot(data=metrics, palette="pastel")
plt.title('Box Plot of Views, Likes, and Dislikes')
plt.xlabel('Engagement Metrics')
plt.ylabel('Count (log scale)')
plt.yscale('log')  # Using a logarithmic scale to better visualize the data spread
plt.grid(True)
plt.show()
```
